# Augmentation with ImageDataGenerator

This is a notebook for nor data only. We will save the picture into cancer_aug folder.

- Image shifts via the width shift range and height shift range arguments.<br>
- Image flips via the horizontal flip and vertical flip arguments.<br>
- Image rotations via the rotation range argument. <br>
- Image brightness via the brightness range argument.<br>
- Image zoom via the zoom range argument.

```python
# example of horizontal shift image augmentation
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
# load the image
path = 'data/small/cancer/'
img = load_img(os.path.join(path,'cancer_DBT-P00107DBT-S05365lcc_33.jpg'))
img.size
```

```python
# convert to numpy array
data = img_to_array(img)
data.shape
```

## Horizontal and Vertical shift

```python
data.shape
```

```python
# expand dimension to one sample
samples = expand_dims(data, 0)
# create image data augmentation generator
datagen = ImageDataGenerator(width_shift_range=[-200,200])
# prepare iterator
it = datagen.flow(samples, batch_size=1)
# generate samples and plot
pyplot.figure(figsize=(5,8))
for i in range(9):
	# define subplot
	pyplot.subplot(3,3 ,1 + i)
	# generate batch of images
	batch = it.next()
	# convert to unsigned integers for viewing
	image = batch[0].astype('uint8')
	# plot raw pixel data
	pyplot.imshow(image)
# show the figure
```

```python
# example of horizontal flip for all
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing.image import save_img
from matplotlib import pyplot
# load the image
path = 'data/small/cancer/'
image_names = os.listdir(path) # 폴더에 있는 이미지들 각 이름들 
# image_names_sample = image_names[0:2]

for i in range(len(image_names)):
    img = load_img(os.path.join(path,image_names[i])) # 각 이미지 이름들 호출 하여 이미지로 저장
    # convert to numpy array
    data = img_to_array(img)
    # expand dimension to one sample
    samples = expand_dims(data, 0) # row 하나더 추가 axis = 0
    
    image = batch[0] # 다시 row를 줄임?
    for j in range(2): # 좌우 반전이 랜덤이라서 2번했음. 
        # create image data augmentation generator
        datagen = ImageDataGenerator(horizontal_flip=True) # 반전이 랜덤임...
        # prepare iterator
        it = datagen.flow(samples, batch_size=1)
        batch = it.next()
        # convert to unsigned integers for viewing
        image = batch[0]
        path1 = 'data/small/c_aug/Hflip'
        save_img(os.path.join(path1,image_names[i].split('.')[0] +'_Hflip_'+ str(j) + '.jpg' ),image)
        print('Total image converted:', len(os.listdir(path1)))
print('Done')
```
