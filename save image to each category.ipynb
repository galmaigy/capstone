```python
# benign first.
df_benign
```

```python
small = pd.read_csv("data/small.csv")
small
# create ID column
```

```python
small['ID'] = small['PatientID']+small['StudyUID']+small['View'] #unique ID
mask = small['Class'] =='benign'
small_b = small.loc[mask, ['ID','Slice','Class']]
small_b
# ID하나에 2개이상의 slice가 존재하네? 음. 그래도 될 듯함. 
```

There are multiple slice for each ID. For example, index 1 and 2, it is same patient with same view but slice 19 and slice 37 confirms benign from multiple images with this ID.

```python
# merge dfs
merged = small_b.merge(df_benign, on ='ID', how = 'left')

merged
```

```python
# slice_1: image number starts with 0, we will use slice_1 for calling later
merged['slice_1'] = merged['Slice']-1
merged_b = merged.copy()
merged_b.head()
```

```python
merged_b.shape
```

```python
#test for 2 lines
merged_b_sample = merged_b.iloc[:3]
merged_b_sample
```

```python
merged_b_sample.iloc[0:3]['slice_1']
```

```python
print('saving images....')
for i in range(merged_b_sample.shape[0]): #총 low 수만큼
    view_series = merged_b_sample.iloc[i] # 각 low를 읽을때, view series라고 명명함. 
    view = view_series["View_x"] # view needed
    image_path = os.path.join("data", view_series["descriptive_path"]) # 각low중에 path 값을 읽어옴
    slice = merged_b_sample.iloc[i]['slice_1'] # slice 정보값을 얻음. 각 row의 slice_1의 값 저장
    ID = merged_b_sample.iloc[i]['ID'] # 각 row의 ID 값 저장. 나중에 파일 명으로 사용함. 
    image = dcmread_image(fp=image_path, view=view) # 저장된 .py 모듈에서 읽어오기. 시간 걸림. 
    img_array = image[slice] # 이미지 배열을 필요한 slice 넘버만 저장
    # reshape the array
    img_arary_r = img_array.reshape((img_array.shape[0], img_array.shape[1],1))
    #save image
    path = 'data/small/benign'
    
    save_img(os.path.join(path,'benign_' + ID +'_'+ str(slice) + '.jpg' ),img_arary_r)
    # it works with sample data set, only saves the image with slice_1 number
```

```python
# 샘플이 성공했음. 이제 전체로 가자. 
merged_b.shape
```

```python
print('saving images....')
for i in range(merged_b.shape[0]):
    view_series = merged_b.iloc[i]
    view = view_series["View_x"] # view needed
    image_path = os.path.join("data", view_series["descriptive_path"])
    # adding slice to extract sliace_1 and ID for naming
    slice = merged_b.iloc[i]['slice_1']
    ID = merged_b.iloc[i]['ID']
    image = dcmread_image(fp=image_path, view=view)
    img_array = image[slice]
    # reshape the array
    img_arary_r = img_array.reshape((img_array.shape[0], img_array.shape[1],1))
    #save image
    path = 'data/small/benign'
    save_img(os.path.join(path,'benign_' + ID +'_'+ str(slice) + '.jpg' ),img_arary_r)
    print('total image counts: ', len(os.listdir(path))) #folder directory
print('Done')
```
